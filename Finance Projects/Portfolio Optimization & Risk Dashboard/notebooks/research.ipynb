```python
from src.risk import (
    portfolio_returns,
    cumulative_returns
)
```

```python
from src.plotting import (
    plot_cumulative_returns,
    plot_drawdown
)
```

```python
portfolio_daily = {}

portfolio_daily["Equal Weight"] = portfolio_returns(
    returns,
    weights
)

portfolio_daily["Minimum Variance"] = portfolio_returns(
    returns,
    min_var_weights
)

portfolio_daily["Maximum Sharpe"] = portfolio_returns(
    returns,
    max_sharpe_weights
)

portfolio_daily["Risk Parity"] = portfolio_returns(
    returns,
    risk_parity_weights
)
```

```python
portfolio_daily = pd.DataFrame(
    portfolio_daily
)
```

```python
cumulative = (
    1 + portfolio_daily
).cumprod()


plot_cumulative_returns(
    cumulative
)
```

```python
drawdown = (
    cumulative /
    cumulative.cummax()
) - 1


plot_drawdown(
    drawdown
)
```
