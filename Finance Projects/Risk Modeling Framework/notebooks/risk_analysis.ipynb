```python
import sys
import os

sys.path.append(
    os.path.abspath("../")
)

from src.data import load_prices
from src.returns import (
    calculate_returns,
    calculate_portfolio_returns
)
```

```python
prices = load_prices(
    "../data/raw/prices.csv"
)

prices.head()
```

```python
returns = calculate_returns(
    prices
)

returns.head()
```

```python
n_assets = prices.shape[1]

weights = [
    1/n_assets
] * n_assets


sum(weights)
```

```python
portfolio_returns = calculate_portfolio_returns(
    returns,
    weights
)

```

```python
portfolio_returns.describe()
```

```python
from src.historical_var import (
    historical_var,
    historical_var_amount
)
```

```python
portfolio_value = 100000

var95 = historical_var(
    portfolio_returns,
    confidence=0.95
)

var95_amount = historical_var_amount(
    portfolio_returns,
    portfolio_value,
    confidence=0.95
)


print("Historical VaR 95%:", var95)
print("Historical VaR 95% (€):", var95_amount)


var99 = historical_var(
    portfolio_returns,
    confidence=0.99
)

print("Historical VaR 99%:", var99)
```
