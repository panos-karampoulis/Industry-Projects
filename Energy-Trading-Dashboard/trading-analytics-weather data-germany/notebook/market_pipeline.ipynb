```python
from meteostat import Hourly
from datetime import datetime
```

```python
start = datetime(2023,1,1)
end = datetime(2023,12,31,23)

weather = Hourly(
    "10637",
    start,
    end
).fetch()

weather.head()
```

```python
weather = weather.reset_index()
```

```python
weather.head()
```

```python
weather = weather[
    [
        "time",
        "temp",
        "rhum",
        "wspd",
        "prcp"
    ]
].copy()
```

```python
weather.rename(
    columns={
        "time": "Datetime",
        "temp": "Temperature",
        "rhum": "Humidity",
        "wspd": "WindSpeed",
        "prcp": "Rain"
    },
    inplace=True
)
```

```python
weather.isna().sum()
```

```python
market_weather = prices.merge(
    weather,
    on="Datetime",
    how="inner"
)
```

```python
market_weather.shape
```

```python
market_weather.to_csv(
    "germany_market_weather_2023.csv",
    index=False
)
```

```python
#create time features
market_weather["Hour"] = market_weather["Datetime"].dt.hour
market_weather["Day"] = market_weather["Datetime"].dt.day
market_weather["Month"] = market_weather["Datetime"].dt.month
market_weather["Weekday"] = market_weather["Datetime"].dt.day_name()
```

```python
#weekend flag
market_weather["Weekend"] = (
    market_weather["Datetime"]
    .dt.weekday
    >= 5
).astype(int)
```

```python
#season (for powerbi)
def season(month):
    if month in [12,1,2]:
        return "Winter"
    elif month in [3,4,5]:
        return "Spring"
    elif month in [6,7,8]:
        return "Summer"
    else:
        return "Autumn"


market_weather["Season"] = (
    market_weather["Month"]
    .apply(season)
)
```

Peak / Off-Peak Hours

This is important in the electricity market.

We define:

Peak:

08:00 - 20:00

Off-peak:

remaining hours

```python
market_weather["Peak_Hour"] = (
    market_weather["Hour"]
    .between(8,20)
).astype(int)
```

```python
#price changes
market_weather["Price_Change"] = (
    market_weather["Price"]
    .diff()
)
```

```python
market_weather["Price_Return"] = (
    market_weather["Price"]
    .pct_change()
)
```

```python
#24-hour moving average
market_weather["Price_MA_24h"] = (
    market_weather["Price"]
    .rolling(24)
    .mean()
)
```

```python
#24-hour volatility
market_weather["Volatility_24h"] = (
    market_weather["Price"]
    .rolling(24)
    .std()
)
```

```python
#Price Spike Detection
market_weather["Spike_Flag"] = (
    market_weather["Price"]
    >
    market_weather["Price_MA_24h"]
    +
    2 * market_weather["Volatility_24h"]
).astype(int)
```

```python
#Negative Price Indicator
market_weather["Negative_Price"] = (
    market_weather["Price"] < 0
).astype(int)
```

```python
#Wind category
market_weather["Wind_Category"] = pd.cut(
    market_weather["WindSpeed"],
    bins=[0,10,20,100],
    labels=["Low","Medium","High"]
)
```

```python
#Heating Degree Days for energy demand
market_weather["HDD"] = (
    18 - market_weather["Temperature"]
).clip(lower=0)
```

```python
#Cooling Degree Days
market_weather["CDD"] = (
    market_weather["Temperature"] - 22
).clip(lower=0)
```

Trading Simulation

We will create a simple yet professional energy trading strategy.

We won't be making a "magic prediction."

We will implement a strategy used for portfolio demonstration:

Mean Reversion Strategy

Concept:

When the price is very low:

→ we buy (Long)

When the price rises back up:

→ we close the position
Signal

We will create:

Price < MA - 1σ → BUY
Price > MA + 1σ → SELL

We will add:

Position
Trade Signal
PnL
Cumulative PnL

```python
#market signal
market_weather["Signal"] = 0
```

```python
#Buy signal:
market_weather.loc[
    market_weather["Price"] <
    (
        market_weather["Price_MA_24h"]
        -
        market_weather["Volatility_24h"]
    ),
    "Signal"
] = 1
```

```python
#Exit signal:
market_weather.loc[
    market_weather["Price"] >
    market_weather["Price_MA_24h"],
    "Signal"
] = -1
```

```python
import numpy as np

#position creation
market_weather["Position"] = (
    market_weather["Signal"]
    .replace(-1, 0)
    .replace(0, np.nan)
    .ffill()
    .fillna(0)
)
```

```python
#PnL
market_weather["PnL"] = (
    market_weather["Position"].shift(1)
    *
    market_weather["Price"].diff()
)
```

```python
#Cumulative PnL
market_weather["Cumulative_PnL"] = (
    market_weather["PnL"]
    .cumsum()
)
```

```python
#results
market_weather[
    [
        "Datetime",
        "Price",
        "Signal",
        "Position",
        "PnL",
        "Cumulative_PnL"
    ]
].tail(20)
```

```python
#KPIs
total_pnl = market_weather["PnL"].sum()

total_trades = (
    market_weather["Signal"]
    .abs()
    .sum()
)

print("Total PnL:", round(total_pnl,2))
print("Trades:", total_trades)
```

```python
#We create real shifts in position.
market_weather["Position_Change"] = (
    market_weather["Position"]
    .diff()
)
```

```python
entries = (
    market_weather["Position_Change"] == 1
).sum()

exits = (
    market_weather["Position_Change"] == -1
).sum()

print("Entries:", entries)
print("Exits:", exits)
```

```python
#Risk Metrics
daily_pnl = (
    market_weather
    .set_index("Datetime")
    ["PnL"]
    .resample("D")
    .sum()
)
```

```python
sharpe = (
    daily_pnl.mean()
    /
    daily_pnl.std()
    *
    np.sqrt(252)
)

print("Sharpe Ratio:", round(sharpe,2))
```

```python
cumulative = daily_pnl.cumsum()
```

```python
running_max = cumulative.cummax()

drawdown = cumulative - running_max

max_drawdown = drawdown.min()

print(
    "Max Drawdown:",
    round(max_drawdown,2)
)
```

```python
VaR_95 = np.percentile(
    daily_pnl,
    5
)

print(
    "VaR 95%:",
    round(VaR_95,2)
)
```

Our dataset shows that no exit was ever triggered after the initial entry.

This often happens when using a simple threshold in volatile markets.

Refining the strategy

We will make the rule more professional:

Entry

We go long when:

Price < MA24 − 1.5 × Volatility

(only truly extreme events)

Exit

We exit when:

the price returns above the MA24

or

we hold the position for a maximum of 24 hours

(very important in trading)

```python
#new Strategy Engine
market_weather["Position"] = 0
```

```python
position = 0
holding_hours = 0

positions = []

for i, row in market_weather.iterrows():

    price = row["Price"]
    ma = row["Price_MA_24h"]
    vol = row["Volatility_24h"]

    if position == 0:

        if price < ma - 1.5*vol:
            position = 1
            holding_hours = 0

    else:

        holding_hours += 1

        if (
            price > ma
            or holding_hours >= 24
        ):
            position = 0
            holding_hours = 0

    positions.append(position)


market_weather["Position"] = positions
```

```python
#new PnL
market_weather["PnL"] = (
    market_weather["Position"].shift(1)
    *
    market_weather["Price"].diff()
)


market_weather["Cumulative_PnL"] = (
    market_weather["PnL"].cumsum()
)
```

```python
#new KPIs
print(
    "Total PnL:",
    round(market_weather["PnL"].sum(),2)
)


print(
    "Trading Hours:",
    (market_weather["Position"] != 0).sum()
)
```

```python
market_weather["Position"].diff().value_counts()
```

```python
#trade events
market_weather["Position_Change"] = (
    market_weather["Position"].diff()
)
```

```python
entries = market_weather[
    market_weather["Position_Change"] == 1
][
    [
        "Datetime",
        "Price"
    ]
].copy()


exits = market_weather[
    market_weather["Position_Change"] == -1
][
    [
        "Datetime",
        "Price"
    ]
].copy()
```

```python
trades = pd.DataFrame()

trades["Entry_Time"] = entries["Datetime"].values
trades["Entry_Price"] = entries["Price"].values

trades["Exit_Time"] = exits["Datetime"].values
trades["Exit_Price"] = exits["Price"].values
```

```python
trades["PnL"] = (
    trades["Exit_Price"]
    -
    trades["Entry_Price"]
)
```

```python
trades["Holding_Hours"] = (
    (
        pd.to_datetime(trades["Exit_Time"])
        -
        pd.to_datetime(trades["Entry_Time"])
    )
    .dt.total_seconds()
    /
    3600
)
```

```python
trades.head()
```

```python
trades.describe()
```

```python
trades["Trade_ID"] = range(1, len(trades)+1)
```

```python
cols = [
    "Trade_ID",
    "Entry_Time",
    "Exit_Time",
    "Entry_Price",
    "Exit_Price",
    "PnL",
    "Holding_Hours"
]

trades = trades[cols]
```

```python
#win/loss ratio
wins = (trades["PnL"] > 0).sum()
losses = (trades["PnL"] <= 0).sum()

win_rate = wins / len(trades)

print("Trades:", len(trades))
print("Wins:", wins)
print("Losses:", losses)
print("Win Rate:", round(win_rate*100,2), "%")
```

```python
#Average Trade
avg_trade = trades["PnL"].mean()

print(
    "Average Trade PnL:",
    round(avg_trade,2)
)
```

```python
#Best / Worst Trade
print(
    "Best Trade:",
    round(trades["PnL"].max(),2)
)

print(
    "Worst Trade:",
    round(trades["PnL"].min(),2)
)
```

```python
#Strategy Summary Table
strategy_summary = pd.DataFrame({
    "Metric": [
        "Total Trades",
        "Total PnL",
        "Average Trade",
        "Win Rate",
        "Best Trade",
        "Worst Trade"
    ],
    "Value": [
        len(trades),
        trades["PnL"].sum(),
        trades["PnL"].mean(),
        win_rate,
        trades["PnL"].max(),
        trades["PnL"].min()
    ]
})

strategy_summary
```
