```python
import yfinance as yf
import pandas as pd

# Download Brent Oil prices
brent = yf.download(
    "BZ=F",
    start="2022-01-01",
    end="2026-01-01",
    auto_adjust=True
)

# Keep only Close price
brent = brent[['Close']]

# Rename column
brent.columns = ['Brent_Price']

# Convert index to column
brent = brent.reset_index()

# Preview
print(brent.head())
print()
print(brent.tail())
print()
print(brent.shape)
```

```python
brent.to_csv('brent_prices.csv', index=False)
```

```python
#TTF Natural Gas
import yfinance as yf
import pandas as pd

# TTF Natural Gas Futures
ttf = yf.download(
    "TTF=F",
    start="2022-01-01",
    end="2026-01-01",
    auto_adjust=True
)

ttf = ttf[['Close']]
ttf.columns = ['TTF_Price']
ttf = ttf.reset_index()

print(ttf.head())
print()
print(ttf.tail())
print()
print(ttf.shape)
```

```python
#EUA Carbon Prices
import yfinance as yf
import pandas as pd

# EUA Carbon Futures
eua = yf.download(
    "CO2.L",
    start="2022-01-01",
    end="2026-01-01",
    auto_adjust=True
)

print(eua.head())
print()
print(eua.shape)
```

```python
#data cleaning EUA
# keep only Close
eua = eua[['Close']]

# rename column
eua.columns = ['EUA_Price']

# transform Date to normal column
eua = eua.reset_index()

print(eua.head())
print()
print(eua.shape)
```

```python
eua.to_csv('eua_prices.csv', index=False)
from google.colab import files
files.download('eua_prices.csv')
```

```python
import yfinance as yf
import pandas as pd

# European energy proxy index
power = yf.download(
    "ENR.DE",   # European energy sector proxy (Germany utilities)
    start="2022-01-01",
    end="2026-01-01",
    auto_adjust=True
)

power = power[['Close']].reset_index()
power.columns = ['Date', 'Power_Index']

power.head()
```

```python
power['Country'] = 'EU'
power.head()
```

```python
#Merge in ONE market table
# ensure Date is datetime
brent['Date'] = pd.to_datetime(brent['Date'])
ttf['Date'] = pd.to_datetime(ttf['Date'])
eua['Date'] = pd.to_datetime(eua['Date'])
power['Date'] = pd.to_datetime(power['Date'])

# merge step by step
df = brent.merge(ttf, on='Date', how='inner')
df = df.merge(eua, on='Date', how='inner')
df = df.merge(power[['Date','Power_Index']], on='Date', how='inner')

df.head()
```

```python
df.to_csv("energy_master_dataset.csv", index=False)
```

```python
df_clean = df[[
    'Date',
    'Brent_Price',
    'TTF_Price',
    'EUA_Price',
    'Power_Index'
]]
```

```python
df_clean.to_csv("energy_prices_clean.csv", index=False)
```
